```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns


df = pd.read_csv(r"C:\Users\kalai\Downloads\updated_student_dataset.csv")


print(df.shape)  
print(df.dtypes)
print(df.isnull().sum())
```

```python
for col in df.select_dtypes(include='number').columns:
    print(f"Column: {col}")
    print(f"Max: {df[col].max()}")
    print(f"Min: {df[col].min()}")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    print(f"Mode: {df[col].mode().values}\n")
```

```python
for col in df.columns:
    print(f"Most frequent value in '{col}': {df[col].mode().values}")
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numeric Features")
plt.show()
```

```python
sns.boxplot(data=df, x='class_12_percent')
plt.title("Boxplot for Outlier Detection")
plt.show()

sns.boxplot(data=df, x='jee_main_score')  
plt.title("Boxplot for Outlier Detection")
plt.show()

sns.boxplot(data=df, x='jee_advanced_score')  
plt.title("Boxplot for Outlier Detection")
plt.show()

sns.boxplot(data=df, x='mock_test_score_avg')  
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
freq = df['coaching_institute'].value_counts()

freq.plot(kind='bar', color='skyblue')
plt.xlabel('Coaching Institutes')
plt.ylabel('Frequency')
plt.title('Frequency of Coaching Institutes')
plt.show()
```
