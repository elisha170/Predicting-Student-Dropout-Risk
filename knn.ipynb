```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv(r"C:\Users\kalai\Downloads\updated_student_dataset.csv")
```

```python
categorical_cols = df.select_dtypes(include=['object', 'string']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
X = df.drop('dropout_yes', axis=1)
y = df['dropout_yes'] 
```

```python
print(y.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

```python
X_train_res, y_train_res = X_train, y_train
```

```python
best_k = 1
best_score = 0.0

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k, weights='distance')
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    print(f"k={k}, Accuracy: {score}")
    if score > best_score:
        best_score = score
        best_k = k

print(f"\nBest k found: {best_k} with Accuracy: {best_score}")
```

```python
final_knn = KNeighborsClassifier(n_neighbors=best_k, weights='distance')
final_knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = final_knn.predict(X_test)

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
